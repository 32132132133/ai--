```python
# 실습

### 붓꽃 데이터 분석

#### 데이터 읽어와서 확인하기

# 데이터 파일 읽어오기
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# 데이터 로드
df = sns.load_dataset('iris')
df

# 데이터의 기본 정보 출력
df.info()

# 데이터의 기초 통계량 출력
df.describe()

# 품종별 개수 구하기
df['species'].value_counts()

# 붓꽃 데이터 전처리

# 결측치 확인하기
df.isna().sum()

# 중복 데이터 확인하기
df.duplicated()  # 각 열마다 중복 데이터 여부를 True/False로
df[df.duplicated()]  # 몇번 인덱스의 어느 행인지 중복 데이터에 해당하는 것을 데이터프레임의 형태로 확인 가능

# 중복 데이터 모두 확인하기
idx = (df['sepal_length'] == 5.8) & (df['petal_width'] == 1.9)  # 이렇게 찾아낸 중복 데이터 값의 인덱스값을 찾아낸 후
df.loc[idx, :]  # 해당 조건에 맞는 행을 모두 표시

# 중복 데이터 삭제하기
df2 = df.drop_duplicates()  # df2에 중복 데이터 제거한 걸 덮어씀
df2[df2.duplicated()]

# 중복 데이터 모두 확인하기
idx = (df['sepal_length'] == 5.8) & (df['petal_width'] == 1.9)  # 찾아낸 중복 데이터 값의 각 요소 길이가 이거였으므로, 이 값을 갖는 인덱스 값을 모두 찾고
df.loc[idx, :]  # 행은 그렇게 찾아낸 모든 인덱스값, 열은 상관없으므로 그냥 모든 열 보이도록 : 해놓으면 해당 행 레이블과 해당 열 레이블의 데이터를 모두 확인할 수 있음.

# 중복 데이터 모두 확인하기
df[df.duplicated()]

# 붓꽃 데이터 그룹핑

# 품종 열을 기준으로 합계 구하기
speciessum = df.groupby('species').sum()
speciessum  # 이렇게 하면 품종 열을 기준으로 각 칼럼의 합계가 구해진다

# 품종 열을 기준으로 평균 구하기
speciesmean = df.groupby('species').mean()
speciesmean  # 이렇게 하면 품종 열을 기준으로 각 칼럼의 평균이 구해진다

# 붓꽃 데이터 시각화

# 막대그래프 그리기
## 꽃받침 길이의 시각화
df['index'] = df.index  # 데이터프레임에 인덱스 칼럼이 없는 경우 임의로 추가해주기
x = df['index']
plt.bar(x, df['sepal_length'])
plt.show()

## 붓꽃 종류별 꽃받침 길이의 평균에 대한 시각화
sns.barplot(x='species', y='sepal_length', data=df)
plt.show()

# 히스토그램 그래프 그리기: 각 특징 4가지 값들의 빈도수에 대한 시각화
fig, axs = plt.subplots(2, 2)
axs[0, 0].hist(df['sepal_length'])
axs[0, 0].set_title('Sepal Length')
axs[0, 1].hist(df['sepal_width'])
axs[0, 1].set_title('Sepal Width')
axs[1, 0].hist(df['petal_length'])
axs[1, 0].set_title('Petal Length')
axs[1, 1].hist(df['petal_width'])
axs[1, 1].set_title('Petal Width')
plt.show()

# 산점도 그래프 그리기: 꽃받침의 길이와 꽃잎의 길이의 산점도를 시각화
sns.scatterplot(x='sepal_length', y='petal_length', hue='species', data=df)
plt.show()
```
